```python
# Im not handling missing values of those columns that are unnecessory, cause noise while having low relevence
```

```python
df = df[['Series_Title', 'IMDB_Rating', 'Genre', 'No_of_Votes', 'Released_Year', ]]
```

```python
df.columns = ['Title', 'Ratings', 'Genre', 'Votes', 'Year']
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
# Calculating Mean, Median, Mode of Ratings
print("Rating Summary:")
print("Mean:", df['Ratings'].mean())
print("Median:", df['Ratings'].median())
print("Mode:", df['Ratings'].mode()[0])
```

```python
# Calculating Mean, Median, Mode of Votes
print("Votes Summary:")
print("Mean:", df['Votes'].mean())
print("Median:", df['Votes'].median())
print("Mode:", df['Votes'].mode()[0])
```

```python
# Visualizing Rating Distribution
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Ratings'], bins=20, kde=True, color='skyblue')
plt.title('Distribution of IMDB Movie Ratings')
plt.xlabel('Ratings')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(y=df['Ratings'], color='lightblue')
plt.title('Boxplot of IMDB Movie Ratings')
plt.show()
```

```python
top_movies = df.sort_values(by=['Ratings', 'Votes'], ascending=False).head(10)
top_movies[['Title', 'Ratings', 'Votes']]
```

```python
# Analysis: Average ratings by genre
df['Ranking_by_genre'] = df['Genre'].apply(lambda x: x.split(',')[0])

genre_ratings = df.groupby('Ranking_by_genre')['Ratings'].mean().sort_values(ascending=False)
genre_ratings
```

```python
# Here's the Visualize average rating
plt.figure(figsize=(10,5))
sns.barplot(x=genre_ratings.index, y=genre_ratings.values, palette='viridis')
plt.xticks(rotation=45)
plt.ylabel('Average Rating')
plt.title('Average IMDB Rating by Genre')
plt.show()
```

```python
df[['Title', 'Ratings', 'Votes', 'Ranking_by_genre', 'Year']].to_csv('movies_cleaned.csv', index=False)
```
